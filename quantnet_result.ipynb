```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
base = "https://quantnet.com/tracker/"
GPA=[]
GREQ=[]
GREV=[]
GREAWA=[]
for i in range(1,341):
    if i==1:
        url=base
    else:
        url=base+'page-'+str(i)
    response = requests.get(url)
    if response.status_code == 200:
        results = BeautifulSoup(response.content,'lxml')
        
        main_div=results.find_all('div',  {"class": "structItem-cell structItem-cell--main"})  
        all_div=results.find_all('dl',  {"class": "pairs pairs--inline"})      
        for _ in main_div:  
            nonempty_check=_.find('dt')
            
            
            if nonempty_check:
                check_tag=_.find('dt')
                check=check_tag.get_text().strip('\n').strip()
                
                if check == 'GRE Q':
                    GPA_tag=_.find('dd')            
                    GPA.append(GPA_tag.get_text().strip('\n').strip())
                elif check == 'UGPA':
                    GREQ_tag=_.find('dd')            
                    GREQ.append(GREQ_tag.get_text().strip('\n').strip())
                elif check == 'GRE V':
                    GREV_tag=_.find('dd')            
                    GREV.append(GREV_tag.get_text().strip('\n').strip())
                elif check == 'GRE AWA':
                    GREAWA_tag=_.find('dd')            
                    GREAWA.append(GREAWA_tag.get_text().strip('\n').strip())
            else:
                GPA.append('N/A')
                GREQ.append('N/A')
                GREV.append('N/A')
                GREAWA.append('N/A')
            
            
    else:
        print("Failure")
```

```python
full=[]
for i in range(len(GPA)):
    new=[]
    new.append(GPA[i])
    new.append(GREQ[i])
    new.append(GREV[i])
    new.append(GREAWA[i])
    full.append(new)
    
```
